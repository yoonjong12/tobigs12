# ë‹¤ìŒ ì˜í™” ì‚¬ì´íŠ¸ Scraping
* í•„ìš”í•œ ì •ë³´ëŠ” 2018ë…„, 2019ë…„ ë°ì´í„°ì˜ [ì œëª©, ë„¤í‹°ì¦Œ í‰ì , í‰ë¡ ê°€ í‰ì , ëˆ„ì  ê´€ê°ìˆ˜]

```python
from bs4 import BeautifulSoup as bs
import requests
import json
import pandas as pd

# ì—°ë„ ë³„ ì˜í™”ë°ì´í„°ë¥¼ ë‹´ì„ í´ë˜ìŠ¤
class Movie_of_Year():
    def __init__(self, inst, year):
        self.inst = inst # ê°ì²´ êµ¬ë¶„ìš©
        self.year = year # ì—°ë„
        self.html = None # bs html 
        self.titles = None # ì œëª©
        self.ids = None # Daumì—ì„œ ì •í•œ ì˜í™” id
        self.netizen_score = [] # ê´€ê° ì ìˆ˜
        self.critic_score = [] # í‰ë¡ ê°€ ì ìˆ˜
        self.total_audience = [] # ì´ ê´€ê°ìˆ˜
    
    # Html 
    def setHtml(self):
        headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; InteSl Mac O X 10_10_1) \
                    AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
        # ê³ ì •ëœ ë§í¬ì— ì—°ë„ë§Œ ë¶™ì—¬ì£¼ë©´ ì—°ë„ë³„ top 100 ì˜í™”ëª©ë¡ì— ì ‘ê·¼í•  ìˆ˜ ìˆìŒ
        url = "https://movie.daum.net/boxoffice/yearly?year=" + self.year
        response = requests.get(url, headers = headers)
        if response.status_code == 200:
            print('request comfirmed!')
        self.html = bs(response.text,'html.parser')
        return self
    
    # ì œëª© 
    def setTitles(self):
        movie_list = self.html.findAll('a',{'class':'name_movie #title'})
        # movie_list['href']ì€ ì£¼ì†Œê°’ë“¤ì„ ì£¼ê¸° ë•Œë¬¸ì— 22ë²ˆì§¸ë¶€í„° ë“±ì¥í•˜ëŠ” idë¥¼ ê°€ì ¸ì˜´. idê°€ ì—†ëŠ” ê²½ìš°ì—ëŠ” None ì²˜ë¦¬
        self.ids = [i['href'][22:] if len(i['href'][22:])>1 else 'None' for i in movie_list ]
        self.titles = [i.text for i in movie_list]
        return self
    
    # ì ìˆ˜ 
    def setScore(self):
        # info_gradeë¥¼ find_allí•˜ë©´ ì—°ë„ ë³„ 100ê°œì˜ ì ìˆ˜ ìŒì„ ì°¾ì„ ìˆ˜ ìˆìŒ
        # ê° iteration ë§ˆë‹¤ ì˜í™” í•˜ë‚˜ì˜ ì ìˆ˜ë¥¼ ê°€ì ¸ì˜¤ëŠ” ê²ƒ.
        info_grade = self.html.findAll('span',{'class':'info_grade'})
        for info in info_grade: 
            movie_grades=info.findAll('span',{'class':'num_grade'})
            # ì ìˆ˜ê°€ 6.8ì´ë©´ '06', '.' , '08'ë¡œ ì €ì¥ë˜ì–´ìˆì–´ì„œ ì²˜ë¦¬í•´ì¤Œ
            movie_scores = [str(int(i.text)) for i in movie_grades if i.text.isdigit()]
            # í•­ìƒ 2ë²ˆì§¸ì™€ 4ë²ˆì§¸ movie_scoresê°€ ë„¤í‹°ì¦Œ, í‰ë¡ ê°€ë¡œ ê³ ì •ë˜ì–´ìˆìŒ. í‰ë¡ ê°€ ì ìˆ˜ëŠ” ì—†ì„ ë•Œ ê³µë°±ìœ¼ë¡œ ì²˜ë¦¬ë˜ì–´ìˆìŒ
            netizen = '.'.join(movie_scores[:2])
            self.netizen_score.append(netizen)
            critic = '.'.join(movie_scores[2:4])
            if len(critic) == 0:
                critic = 'None'
            self.critic_score.append(critic)
        return self
    
    # ì´ ê´€ê°ìˆ˜ 
    def setAudience(self):
        # ì˜í™” idë¥¼ linkë’¤ì— ë¶™ì´ë©´ íŠ¹ì´í•˜ê²Œë„ json dataë¥¼ ë°›ì„ ìˆ˜ ìˆìŒ
        # json ë¼ì´ë¸ŒëŸ¬ë¦¬ë¡œ ì‰½ê²Œ ë½‘ì•„ì˜¬ ìˆ˜ ìˆìŒ
        for i in self.ids:
            link = 'https://movie.daum.net/moviedb/main/totalAudience.json?movieId='
            total_link = link + i
            try:
                parser = requests.get(total_link).text
                dic = json.loads(parser)
                total_aud = dic['totalAudience']   
            except:
                total_aud = 'None'
            self.total_audience.append(total_aud)
        return self
    
    # ê·¸ë™ì•ˆ ìŒ“ì•„ì˜¨ attributesë“¤ì„ ê³¨ë¼ì„œ ë¶ˆëŸ¬ì˜¤ê¸°
    def getData(self):
        # ì—°ë„, ì œëª©, í‰ì , ëˆ„ì  ê´€ê°ìˆ˜ë§Œì„ ì›í•¨
        movie_data = self.__dict__.copy()
        del_li = ['html','ids','inst']
        for i in del_li:
            del movie_data[i]
        return movie_data
    
my_objects = []
df = pd.DataFrame()
for i, j in enumerate(['2018','2019']):
    obj = Movie_of_Year(i,j)
    my_objects.append(obj)
    obj.setHtml()
    obj.setTitles()
    obj.setScore()
    obj.setAudience()
    movie_data = obj.getData()
    movie_data = pd.DataFrame(movie_data,index=obj.ids)
    df = pd.concat([df,movie_data])
```
